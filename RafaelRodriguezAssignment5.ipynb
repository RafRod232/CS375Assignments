## <font color='blue'> Question 0. MLP with scikit-learn: Explore outputs for the MNIST dataset [40 pts]</font>

The purpose of this question is to get some practice time with MLP classifiers. We will be using the the MNIST data set.




```python
# Let's first load the MNIST data set.

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split


X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
y = y.astype(int)
X = ((X / 255.) - .5) * 2
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=10000, random_state=123, stratify=y)
```

<br> **Q0-0**: This is the classifier we built in class. Find a setting for *batch_size* and *max_iter*, so that the subsequent fit function **converges**. Converge here is defined as: the training loss does not improve by more than tol=1e-4 for n_iter_no_change=10 consecutive passes over the training set. [10 pts]





```python

mlp_mnist = MLPClassifier(verbose = True, hidden_layer_sizes=(50,), batch_size = 5, \
                          max_iter=50, solver='sgd', activation='logistic',\
                          learning_rate = 'constant', learning_rate_init = 0.001, random_state=1)

#create the model of mlp(multi layer perceptron) classifier
```

```python
mlp_mnist.fit(X_train,y_train)
# fitting the model with the training data
```

```python
predictions = mlp_mnist.predict(X_test)
probability_predictions = mlp_mnist.predict_proba(X_test)
```

**Q0-1**: Write a short function that outputs the **second best** prediction for each point, i.e. the label which gets the second highest probability in the softmax output. [10 pts]

```python
### Your code here
for i in range(len(probability_predictions)):
    arrary = probability_predictions[i].copy()
    arrary.sort()
    secprob = arrary[-2]
    print(f"The point {i} had its second highest probability be {secprob}")
```

**Q0-2**: For each data point misclassified by the model, find if the second best prediction is actually the correct label. Calculate the percentage of missclassified points, that are correctly classified based on the second best prediction. [10 pts]

```python
### Your code here. Also give some comments about your findings
# We are comparing the predictions with the y test
mismatch = 0
index_of_mismatch_ytest = []
for i in range(len(predictions)):
    if(predictions[i]==y_test.iloc[i]):
        continue
    else:
        mismatch += 1
        index_of_mismatch_ytest.append(i)
        #store the label index into an array of y_test
        continue
mismatch_percent = mismatch/10000
print(f"There are {mismatch} labels that are missclassified from y_test meaning that {mismatch_percent}% of the y_pred are missclassified points")
print("Now checking if the second highest predictions is the correct label")
#retrieving index of value
def ind(array, item):
    for idx, val in np.ndenumerate(array):
        if val == item:
            return idx
#print(index_of_mismatch)
sec_prob_corr = 0
for i in index_of_mismatch_ytest:
    arr = probability_predictions[i].copy()
    arr.sort()
    prob2 = arr[-2] #returns the value of the second highest prob
    #prob2 = prob2.astype(int)
    prob2_indx = ind(probability_predictions[i],prob2)
    if int(prob2_indx[0]) == y_test.iloc[i]:
        sec_prob_corr+=1
print(f"The number of correctly classified lables as the second highest probablity is {sec_prob_corr} points")
print(f"The percentage of missclassified points that are correctly classifed as the second highest probability is {sec_prob_corr/mismatch}%")
```
